```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
homes = pd.read_csv('http://files.zillowstatic.com/research/public/Metro/Sale_Prices_Msa.csv')

# This dataset contains the median sale prices for homes in the US by region. See here for more information: https://www.zillow.com/research/data/

print(homes.shape)

homes.head()
```

```
rents = pd.read_csv('http://files.zillowstatic.com/research/public/Metro/Metro_MedianRentalPrice_AllHomes.csv')

# This dataset contains the median rental list prices in the US by region. See here for more information: https://www.zillow.com/research/data/

print(rents.shape)

rents.head()
```

```
# I want to merge these dataframes. However, I need to maintain the distinction between home prices and rent prices.
# To do this, I'll rename the columns in the dataframes.

list(homes)
```

```
home_names = [f'Home {s}' for s in list(homes)]

rent_names = [f'Rent {s}' for s in list(rents)]

home_names
```

```
homes = pd.read_csv('http://files.zillowstatic.com/research/public/Metro/Sale_Prices_Msa.csv', names=home_names, skiprows=1)

rents = pd.read_csv('http://files.zillowstatic.com/research/public/Metro/Metro_MedianRentalPrice_AllHomes.csv', names=rent_names, skiprows=1)
```

```
df = pd.merge(homes, rents, left_on='Home RegionName', right_on='Rent RegionName')

print(df.shape)

df.head()
```

```
# Dropping the redundant columns.

df = df.drop(['Home RegionID', 'Rent RegionName', 'Rent SizeRank'], axis=1)

df = df.rename({'Home RegionName': 'RegionName', 'Home SizeRank': 'SizeRank'}, axis=1)

print(df.shape)

df.head()
```

```
# The homes dataset contains some months that the rents dataset lacks. Since I'm interested in comparing them, I don't need those months.
# Dropping these months is also important so that, when I take the average across all months, I have the same parameters for both homes and rents.

df = df.drop(columns=['Home 2008-03', 'Home 2008-04', 'Home 2008-05', 'Home 2008-06', 'Home 2008-07', 'Home 2008-08',
             'Home 2008-09', 'Home 2008-10', 'Home 2008-11', 'Home 2008-12', 'Home 2009-01', 'Home 2009-02',
             'Home 2009-03', 'Home 2009-04', 'Home 2009-05', 'Home 2009-06', 'Home 2009-07', 'Home 2009-08',
             'Home 2009-09', 'Home 2009-10', 'Home 2009-11', 'Home 2009-12'])

print(df.shape)

df.head()
```

```
# Since there's only one NaN value, it'll be worthwhile to fill it. Prices don't seem to change significantly
# when you go forward or backward a month, so I think bfill or ffill is the best method.

df['Home 2019-01'] = df['Home 2019-01'].fillna(method='bfill', axis=0)

df['Home 2019-01'].isnull().sum()
```

```
# Now to calculate the quotient between home and rent prices for January 2019.
# I think division is better than subtraction here because home prices are so much bigger than rent prices;
# subtraction would be too biased in favor of the home prices.

df['Quotient 2019-01'] = df['Home 2019-01'] / df['Rent 2019-01']

df.head()
```

```
# Now I can achieve my main goal here, which is to average the prices across all months.

homes_region_averages = [np.mean(homes2.iloc[:, i]) for i in range(len(homes2.columns))]
homes_region_averages
```

```
df['Home Averages'] = homes_region_averages
df['Rent Averages'] = rents_region_averages

df.head()
```

```
df['Average Quotients'] = df['Home Averages'] / df['Rent Averages']
df.head()
```

```
# Finally, I want to graph the quotient of home and rent prices in the entire country over time.
# First, to get the US data by itself.

us_homes = homes2[['index', 'United States']]

us_homes.head()
```

```
us_rents = rents3[['index', 'United States']]
us_rents.head()
```

```
# I need to fill the NaN values now so that every month is represented.
# As before, I think ffill and bfill are the best. I use ffill for homes
# because the NaN is at the end, and bfill for rents because the NaNs are at the beginning.

us_homes = us_homes.fillna(method='ffill')
us_homes.isnull().sum()
```

```
us_rents = us_rents.fillna(method='bfill')
us_rents.isnull().sum()
```

```
print(us_homes.shape)
print(us_rents.shape)
```

```
us_rents.head()
```

```
us = pd.concat([us_homes, us_rents], axis=1)

print(us.shape)

us.head()
```

```
us.columns = ['months1', 'homes', 'months2', 'rents']
us.head()
```

```
us['quotients'] = us['homes'] / us['rents']
us.head()
```

```
# It'll be convenient to have the months in datetime format. First I need to get rid of the 'Home' part of the string.

us['months1'] = us['months1'].str.replace('Home', '')
us.head()
```

```
# Now to convert to datetime!

us['months1'] = pd.to_datetime(us['months1'], infer_datetime_format=True, errors='coerce')
us.dtypes
```

```
us.isnull().sum()
```

```
us['months1'].dt.month
```

```
fig, ax = plt.subplots()
ax.plot('months1', 'quotients', data=us)
ax.set_xlabel('Month')
ax.set_ylabel('Quotient')
ax.set_title('Median Home Price Divided by Median Rent Price in the US Over Time')
ax.set_facecolor('#F0F0F0')
plt.show()
```

```
fig, ax = plt.subplots(2)
ax[0].plot('months1', 'homes', data=us)
ax[1].plot('months1', 'rents', data=us)
ax[0].set_ylabel('Median home price')
ax[1].set_ylabel('Median rent price')
ax[1].set_xlabel('Month')
ax[0].set_title('Median Home and Rent Prices Over Time')
ax[0].set_facecolor('#F0F0F0')
ax[1].set_facecolor('#F0F0F0')
plt.show()
```
